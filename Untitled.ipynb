```python
size = len(data) * 18
random_data = np.random.rand(size, 5)

# Create a pandas DataFrame from the numpy array
organized_data = pd.DataFrame(random_data, columns=['Building_index', 'week_of_year', 'day_of_week', 'hour_of_day',"normalized_energy"])
index = 0

build_max_consumption = []
for i in range(19):
    if i == 5:
        build_max_consumption.append(None)
    if i != 5:
        build_max_consumption.append(np.max(data["Bldg%d (kWh)"%(i+1)]))

for i in range(len(data)):
    temp = data.loc[i]
    date_string = temp["Start Time (GMT-0800,PST)"]
    date_obj = datetime.datetime.strptime(date_string, '%m/%d/%y %H:%M')
    week_of_year = date_obj.isocalendar()[1]
    day_of_week = date_obj.weekday()
    hour_of_day = date_obj.hour
    for j in range(19):
        if j != 5:
            value = temp["Bldg%d (kWh)"%(j+1)] / build_max_consumption[j]
            organized_data.loc[index] = [(j+1), week_of_year, day_of_week,hour_of_day, value ]
            index += 1
print(organized_data)
```

```python
# size = len(organized_data) - 3* 18
# tt = organized_data[organized_data["Building_index"] == 1].reset_index()
# size_for_one_building = len(organized_data[organized_data["Building_index"] == 1]) - 3
# print(tt[0:25])
organized_data_training = pd.DataFrame(columns=[ 'day_of_week', 'hour_of_day',"normalized_energy","average_energy","t-3","t-2","t-1","true_value"])
```

```python
for i in range(19):
    if i ==5:
        continue
    tt = organized_data[organized_data["Building_index"] == (i+1)].reset_index()
    size_for_one_building = len(organized_data[organized_data["Building_index"] == (i+1)]) - 3
    temp_data =  np.random.rand(size_for_one_building, 8)
    temp_data = pd.DataFrame(temp_data, columns=[ 'day_of_week', 'hour_of_day',"normalized_energy","average_energy","t-3","t-2","t-1","true_value"])
    for i in range(0,3):
        day_of_week = tt.loc[i,'day_of_week' ]
        hour_of_day = tt.loc[i,'hour_of_day' ]
        normalized_energy = tt.loc[i,'normalized_energy' ]
        average_energy = grouped_by_hour[day_of_week][hour_of_day]
        true_value = normalized_energy - average_energy
        temp_data.iloc[i] = [day_of_week,hour_of_day,normalized_energy,average_energy,0,0,0,true_value]

    for i in range(3, size_for_one_building):
        day_of_week = tt.loc[i,'day_of_week' ]
        hour_of_day = tt.loc[i,'hour_of_day' ]
        normalized_energy = tt.loc[i,'normalized_energy' ]
        average_energy = grouped_by_hour[day_of_week][hour_of_day]
        true_value = normalized_energy - average_energy
        t_minus3 = temp_data.loc[i-3,"true_value"]
        t_minus2 = temp_data.loc[i-2,"true_value"]    
        t_minus1 = temp_data.loc[i-1,"true_value"]
        temp_data.iloc[i] = [day_of_week,hour_of_day,normalized_energy,average_energy,t_minus3,t_minus2,t_minus1,true_value]
    organized_data_training = pd.concat((organized_data_training, temp_data[3:]), axis=0).reset_index(drop=True)
print(organized_data_training)
```

```python
organized_data_training.to_csv('organized_data_training.csv', index=False)
```
